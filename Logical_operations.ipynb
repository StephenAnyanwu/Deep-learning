## Computing Logical Operations Using NN 

Logical operators are used with expressions whose results are logical values, That is, True or False. We are going to use binary numbers to represent these values, True = 1 and False = 0. We will build neural networks that perform basic logical operation

### AND Opreator:
The output is "true" when both inputs are "true." Otherwise, the output is "false." In other words, the output is 1 only when both inputs one AND two are 1.


and.PNG

nn.PNG

$\sum$ : weighted sum $(z)$<p>
$f$ : activation function (sigmoid function $\sigma$)<p>
$z=w_1x_1+w_2x_2+b=\displaystyle\sum^2_{i=1}w_ix_i+b$<p>
$\sigma=\frac{1}{1+e^z}$<p>
$y=\sigma(z)$



```python
import numpy as np 

# Assign values to the parameters
#Let bias b = w_0
w_0 = -30
w_1 = 20
w_2 = 20

# Values of the binary inputs
x_1 = np.array([1, 1, 0, 0])
x_2 = np.array([1, 0, 1, 0])

print(f"Inputs:\n X_1 = {x_1} \n X_2 = {x_2}\n")

# Vector of weighted sum of the inputs
z = np.array([w_1 * x_1 + w_2 * x_2 + w_0 for x_1, x_2 in zip(x_1, x_2)])
print(f"Vector of weighted sum of the inputs: \n z = {z}\n")

# Vector of outputs
y = np.round(1/(1 + np.exp(-z)), 3)
print(f"Outputs X_1 AND X_2: \n {y}")

```

### XOR Operator (Exclusive OR):
The XOR ( exclusive-OR ) operator acts in the same way as the logical "either/or." The output is "true" if either, but not both, of the inputs are "true." The output is "false" if both inputs are "false" or if both inputs are "true." Another way of looking at this operator is to observe that the output is 1 if the inputs are different, but 0 if the inputs are the same. 

image.png

```python
# Assign values to the parameters
# Let bias b = w_0
w_0 = -10
w_1 = 20
w_2 = 20

# Values of the binary inputs
x_1 = np.array([1, 1, 0, 0])
x_2 = np.array([1, 0, 1, 0])

print(f"Inputs:\n X_1 = {x_1} \n X_2 = {x_2}\n")

# Vector of weighted sum of the inputs
z = np.array([w_1 * x_1 + w_2 * x_2 + w_0 for x_1, x_2 in zip(x_1, x_2)])
print(f"Vector of weighted sum of the inputs: \n z = {z}\n")

# Vector of outputs
y = np.round(1/(1 + np.exp(-z)), 3)
print(f"Outputs of X_1 OR X_2: \n {y}")
```

### XNOR Operator:
The XNOR (exclusive-NOR) operator is a combination XOR operator followed by an inverter. Its output is "true" if the inputs are the same, and "false" if the inputs are different. That is, its output is 1 if $x_1$ and $x_2$ are the same.

xnor.PNG

mav.PNG

```python
# Hidden layer parameters
# Let bias b = w_0
w_h01 = -30
w_h11 = 20
w_h12 = 20

w_h02 = 10
w_h21 = -20
w_h22 = -20

# Output layer parameters
w_out01 = -10
w_out11 = 20
w_out21 = 20

# Values of the binary inputs
x_1 = np.array([1, 1, 0, 0])
x_2 = np.array([1, 0, 1, 0])
print(f"Inputs:\n X_1 = {x_1} \n X_2 = {x_2}\n")

a_h1_vec = []
a_h2_vec = []
y = []
for x_1, x_2 in zip(x_1, x_2):
  inputs = np.array([x_1, x_2])
  w_h = np.array([[w_h11, w_h12], [w_h21, w_h22]]) 
  z_h = np.dot(inputs, w_h.T) + np.array([w_h01, w_h02])
  a_h = np.array([1/(1 + np.exp(-z)) for z in z_h])
  w_out = np.array([w_out11, w_out21])
  z_out =  np.dot(a_h, w_out) + w_out01
  a_out = np.round(1/(1 + np.exp(-z_out)), 1)
  a_h1_vec.append(np.round(a_h[0], 1))
  a_h2_vec.append(np.round(a_h[-1], 1))
  y.append(a_out)
print(f"a_h1:\n {np.array(a_h1_vec)}\n")
print(f"a_h2:\n {np.array(a_h2_vec)}\n")
print(f"Outputs of x_1 XNOR x_2 \n {np.array(y)}")
```

### NAND Operator:
The NAND operator operates as an AND operator followed by a NOT operator. It acts in the manner of the logical operation "and" followed by negation. The output is "false" if both inputs are "true." Otherwise, the output is "true."

```python
# Assign values to the parameters
#Let bias b = w_0
w_0 = 30
w_1 = -20
w_2 = -20

# Values of the binary inputs
x_1 = np.array([1, 1, 0, 0])
x_2 = np.array([1, 0, 1, 0])

print(f"Inputs:\n X_1 = {x_1} \n X_2 = {x_2}\n")

# Vector of weighted sum of the inputs
z = np.array([w_1 * x_1 + w_2 * x_2 + w_0 for x_1, x_2 in zip(x_1, x_2)])
print(f"Vector of weighted sum of the inputs: \n z = {z}\n")

# Vector of outputs
y = np.round(1/(1 + np.exp(-z)), 3)
print(f"Outputs X_1 NAND X_2: \n {y}")
```

### (NOT $x_1$) AND (NOT $x_2$):
The output is true if and only if $x_1$ and $x_2$ are false. For the output to be 1, both inputs must be 0. It is the same as NOR operator.

notx.PNG

nre.PNG

```python
# Hidden layer parameters
# Let bias b = w_0
w_h01 = 10
w_h02 = 10
w_h11 = -20
w_h22 = -20

# Output layer parameters
w_out01 = -30
w_out11 = 20
w_out21 = 20

# Values of the binary inputs
x_1 = np.array([1, 1, 0, 0])
x_2 = np.array([1, 0, 1, 0])
print(f"Inputs:\n X_1 = {x_1} \n X_2 = {x_2}\n")

a_h1_vec = []
a_h2_vec = []
y = []
for x_1, x_2 in zip(x_1, x_2):
  z_h1 = w_h11 * x_1 + w_h01
  z_h2 = w_h22 * x_2 + w_h02
  a_h1 = np.round(1/(1 + np.exp(-z_h1)), 3)
  a_h2 = np.round(1/(1 + np.exp(-z_h2)), 3)
  z_out = w_out11 * a_h1 + w_out21 * a_h2 + w_out01
  a_h1_vec.append(int(a_h1))
  a_h2_vec.append(int(a_h2))
  a_out =  np.round(1/(1 + np.exp(-z_out)), 1)
  y.append(a_out)
print(f"a_h1:\n {np.array(a_h1_vec)}\n")
print(f"a_h2:\n {np.array(a_h2_vec)}\n")
print(f"Outputs of (NOT x-1) AND (NOT x_2)\n {np.array(y)}")

```
